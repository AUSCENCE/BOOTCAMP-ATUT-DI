```python
import warnings
from fileinput import filename
warnings.filterwarnings("ignore")
import json
from crewai import Agent, Task, Crew,process
from crewai_tools import BaseTool,SerperDevTool
from  langchain_groq import ChatGroq
from langchain_ollama import ChatOllama
import os
import pandas as pd
from cerberus import Validator



os.environ["OPENAI_API_KEY"] ="NA"

#llm = ChatOllama(model="llama3.1",base_url="http://localhost:11434")
llm_groq = ChatGroq(temperature=0,groq_api_key="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",model_name="llama-3.1-70b-versatile")

class SaveCSVTool(BaseTool):
    name: str ="SaveCSVTool"
    description: str ="Tool to save provided data into a CSV file."

    def _run(self,data:dict, filename: str ="data_sites_touristiques.csv") -> str:
        df = pd.DataFrame(data["sites_touristiques"])
        if(self.validate(data)):
          df.to_csv(filename,index=False)
          return "Data validated -> CSV saved as {filename}"
        else:
          return "Invalid data"

    def _validate(self,data:dict) -> bool:
        content_schema = {
            "sites_touristiques": {
                "type": "list",
                "schema": {
                    "type": "dict",
                    "schema": {
                        "nom": {"type": "string"},
                        "adresse": {"type": "string"},
                        "coordonnees_gps": {"type": "string"},
                        "description": {"type": "string"},
                        "url": {"type": "string"},
                    },
                    "required": ["nom", "adresse", "coordonnees_gps", "description", "url"],
                    "additionalProperties": False,
                },
                "required": ["sites_touristiques"],
                "additionalProperties": False,
                }
            }
        validator = Validator(content_schema)
        return validator.validate(data)

agent= Agent(
    role="Search Specialist",
    goal="Trouve et liste tous les sites touristiques du Bénin",
    backstory="Tu es un expert en recherche des sites touristiques du Bénin.Ta tâche est d'identifier et de lister tous les sites touristiques du Bénin.",
    verbose=True,
    llm=llm_groq,
    tools = [SaveCSVTool(result_as_answer=True)]
)

task= Task(
    description=(
        "rechercher et lister tous les sites touristiques du Bénin"
        "Inclure l'adresse du site et les coordonnées GPS sous forme de lien cliquable"
        "Inclure leur description complète"
        "Inclure les urls"
    ),
    expected_output="un objet json comme réponse avec pour clé 'sites_touristiques",
    agent=agent
)

crew = Crew(
    agents= [agent],
    tasks= [task],
    verbose=True,
)


result = crew.kickoff(inputs={"search_query":"Tous les sites touristiques du Bénin"})


print(result)
```
